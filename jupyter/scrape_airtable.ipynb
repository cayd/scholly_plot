# Scrape the table

```python
import requests
import re
from datetime import date
import pandas as pd
```

```python
base_id = "appdbVBj7AnaiBL00"
table_name = "Berkeley%202021%20Scholarships "
url = "https://api.airtable.com/v0/" + base_id + "/" + table_name

api_key = "keyqosr8hSKDywgVJ"
headers = {"Authorization": "Bearer " + api_key}
params = ()

response = requests.get(url, params=params, headers=headers)
airtable_response = response.json()

airtable_records = airtable_response['records']

airtable_records
```

```python
df_list = []

for record in airtable_records:
    if ("LSAT" not in record['fields'] or 
       "GPA" not in record['fields'] or 
       ('Initial Scholarship Offer (Total over 3 years)' not in record['fields'] and 
        'Final Scholarship Offer' not in record['fields'])):
        continue
    
    df_list.append(pd.DataFrame({
        "Admission Date": [date.today()],
        "LSAT": [int(record['fields']['LSAT'])],
        "GPA": [float(re.search('\d\.\d+', record['fields']['GPA']).group()) 
                if re.search('\d\.\d+', record['fields']['GPA'])
                else 0],
        "URM?": ["Yes" if "URM?" in record['fields'] else "No"],
        "Date Scholarship Received": [date.today()],
        "Scholarship": [record['fields']['Final Scholarship Offer'] 
                        if 'Final Scholarship Offer' in record['fields']
                        else record['fields']['Initial Scholarship Offer (Total over 3 years)']],
        "Named Scholarship?": ["No"],
        "Conditional Scholarship?": ["No"]
    }))
    
df = pd.concat(df_list)

df
```
